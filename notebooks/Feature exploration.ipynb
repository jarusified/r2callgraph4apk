```python
import os
import pandas as pd
```

```python
_pwd = os.path.abspath("..")
_data_dir = os.path.join(_pwd, "data/features")
_features_csv = os.path.join(_data_dir, "malgenome-215-dataset-1260malware-2539-benign.csv")
_categories_csv = os.path.join(_data_dir, "dataset-features-categories.csv")
```

```python
features_df = pd.read_csv(_features_csv)
categories_df = pd.read_csv(_categories_csv)
```

```python
features_df
```

```python
b_df = features_df.loc[features_df['class'] == 'S']
m_df = features_df.loc[features_df['class'] == 'B']
```

```python
# TODO: avoid the lookup warning
def df_count_values_rowwise(df, val):
    return df.eq(1).sum(axis=1)
```

```python
m_df['factors'] = df_count_values_rowwise(m_df, 1)
b_df['factors'] = df_count_values_rowwise(b_df, 1)

print(f"Number of malicious factors that influence: {m_df['factors'].unique()}")
print(f"Number of benign factors that influence: {b_df['factors'].unique()}")
```
