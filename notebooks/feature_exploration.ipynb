```python
def request_apk_andro(api_key, sha256, data_dir):
    import requests
    url = "https://androzoo.uni.lu/api/download?"

    packaged_url = url + 'apikey=' + api_key + '&sha256=' + sha256
    print("Requesting", packaged_url)
    r = requests.get(packaged_url)
    
    filename = os.path.join(data_dir, sha256 + ".apk")
    with open(filename, 'wb') as fd:
        for chunk in r.iter_content(chunk_size=128):
            fd.write(chunk)
```

```python
API_KEY=open(os.path.join(_pwd, 'ANDROZOO_API_KEY.txt'),"r").read()
print(API_KEY)
```

```python
def get_all_malicious(benign_sha256, data_dir, folder_name):
    malicious_sha_list = apk_df.loc[apk_df['ORIGINAL_APP'] == benign_sha256]['PIGGYBACKED_APP']
    
    benign_folder_path = os.path.join(os.path.join(data_dir, folder_name), "benign")
    print(benign_folder_path)
    os.makedirs(benign_folder_path)
    request_apk_andro(API_KEY, str(benign_sha256), benign_folder_path)
    
    malicious_folder_path = os.path.join(os.path.join(data_dir, folder_name), "malicious")
    print(malicious_folder_path)
    os.makedirs(malicious_folder_path)
    for malicious_sha256 in malicious_sha_list:
        request_apk_andro(API_KEY, str(malicious_sha256), malicious_folder_path)
```

```python
for idx, sha256 in enumerate(apk_df['ORIGINAL_APP'].unique().tolist()[:10]):
    get_all_malicious(sha256, _data_dir, f"app_{idx}")
```
